```python
import pandas as pd
import io


file_name = list(uploaded.keys())[0]
# read the uploaded csv file into a pandas dataframe, specifyingthe 'encoding as 'iso-8859-1

df = pd.read_csv(io.BytesIO(uploaded[file_name]),encoding='ISO-8859-1')

# print the first 10 rowsof the data
df.head(10)
```

```python
# drop unnecessary columns
df= df.drop(columns=['Row ID','Postal Code'])
```

```python
df.describe()
```

```python
# convert data columns to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
print(df.info())
```

```python
# Convert categorical columns to 'category' data type
columns_to_convert = ['Category','Product Name','Region','State','Ship Mode','Order ID','Customer ID','Customer Name','Segment','Country','City','State']
for col in columns_to_convert:
    df[col] = df[col].astype('category')
print(df.info())
```

```python
# What are the most sold categories ?
most_sold_categories = df.groupby('Category',observed=False)['Quantity'].sum().sort_values(ascending=False)
print(most_sold_categories)
```

```python
# What are the most sold product?
most_sold_products = df.groupby('Product Name',observed=False)['Quantity'].sum().sort_values(ascending=False)
print(most_sold_products.head(10))
```

```python
# what categories generate the highest profit?
highest_profit_categories = df.groupby('Category',observed=False)['Profit'].sum().sort_values(ascending=False)
print(highest_profit_categories)
print("\n" + "="*30 + "\n")
```

```python
# what products generate the highest profit?
highest_profit_products = df.groupby('Product Name',observed=False)['Profit'].sum().sort_values(ascending=False)
print(highest_profit_products.head(10))
print("\n" + "="*30 + "\n")
```

```python
highest_profit_regions = df.groupby('Region',observed=False)['Profit'].sum().sort_values(ascending=False)
print(highest_profit_regions)
print("\n" + "="*30 + "\n")
```

```python
# Are there any loss-making products?
loss_making_products = df[df['Profit']<0].groupby('Product Name',observed=False)['Profit'].sum().sort_values(ascending=True)
print(loss_making_products.head(10))
print("\n" + "="*30 + "\n")
```

```python
# what is the relationship between discount and profit?
correlation = df['Discount'].corr(df['Profit'])
print(f":{correlation: .2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['Discount'], df['Profit'], alpha=0.5, color='gray')
plt.title('Discount vs. Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
average_profit_by_discount = df.groupby('Discount')['Profit'].mean().sort_index()
print(average_profit_by_discount)
```

```python
large_discount_losses = df[(df['Discount']>=0.5)&(df['Profit']<0)]
print(large_discount_losses.head(10))
print("\n" + "="*30 + "\n")
print(f"\n Discount>=0.5: {len(large_discount_losses)}")
```

```python
import matplotlib.pyplot as plt
# How are sales distrubted over time? are they any seasons or month with higher sales?
df ['Order Month'] = df['Order Date'].dt.to_period('M')
salse_over_time = df.groupby('Order Month')['Sales'].sum()
print(salse_over_time)
```

```python
plt.figure(figsize=(10,4))
salse_over_time.plot(kind='line')
plt.title('sales_over_time')
plt.xlabel('monthe_and_year')
plt.ylabel('sales')
plt.grid(True)
plt.show()
```

```python
# what is the performance of different shipping modes in terms of sales and profit?
ship_mode_performance = df.groupby('Ship Mode',observed=False)[['Sales','Profit']].sum()
print(ship_mode_performance)
print("\n" + "="*30 + "\n")
```

```python
# Is there a relationship between delevery time and profitability or customer satisfaction?
df['Delivery Time'] = (df['Ship Date']- df['Order Date']).dt.days
average_profit_by_delivery_time = df.groupby('Delivery Time')['Profit'].mean()
print(average_profit_by_delivery_time)
```

```python
# identify the most loyal customers based on the number of order
customer_loyalty_by_orders = df.groupby('Customer ID',observed=False)['Order ID'].nunique().sort_values(ascending=False)
print(customer_loyalty_by_orders.head(10))
print("\n" + "="*30 + "\n")
```

```python
# Save the processed data to a csv file
file_path_csv = 'processed_superstore_sales_dataset.csv'
df.to_csv(file_path_csv,index=False,encoding='utf-8')
print(f"{file_path_csv}")
```
